```python
import random
from matplotlib import pyplot as plt 
```

```python
celcius = []

for i in range(0,30):
    n = random.randint(5,40)
    celcius.append(n)

print(celcius)
```

```python
days = []

for i in range(1,31):
    days.append(i)
    
print(days)
    
```

```python
plt.plot(days, celcius, color='green', marker='o') 
plt.title("Mayıs Ayı Hava Sıcaklığı Değişimi")
plt.xlabel("Günler")
plt.ylabel("Sıcaklık (Celcius)")
plt.show()
```

```python
def average(array):
    return sum(array) / len(array)
```

```python
ortalama = average(celcius)

print(ortalama)
```

```python
import numpy
```

```python
standart_sapma = numpy.std(celcius)

print(standart_sapma)
```

```python
import statistics

varyans = statistics.variance(celcius)

print(varyans)
```

```python
en_yuksek = max(celcius)

print(en_yuksek)
```

```python
en_dusuk = min(celcius)

print(en_dusuk)
```
